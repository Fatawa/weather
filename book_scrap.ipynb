```python
import csv
from bs4 import BeautifulSoup
import requests
```

```python
def convert(num):
    rating= {'One':1 , "Two":2, "Three":3, "Four":4, "Five":5}
    return rating[num]
books =[]
```

```python
response = requests.get("https://books.toscrape.com/")
soup = BeautifulSoup(response.text)
categories_a = soup.find('div',attrs={'class':"side_categories"}).find('ul').find_all('a')
for category_a in categories_a:
    category_name = category_a.text.strip()
    category_url ="https://books.toscrape.com/"+category_a.attrs['href']
    response = requests.get(category_url)
    soup = BeautifulSoup(response.text)
    articles = soup.find_all('article',attrs={'class':'product_pod'})
    for article in articles:
        title = article.find('h3').find('a').text
        price = float(article.find('p',attrs={'class':'price_color'}).text.split("£")[1])
        rating = article.find('p',attrs={'class':'star-rating'}).attrs['class'][1]
        rating = convert(rating)
        book_data = {'book_name': title, 'price': price, 'rating': rating, 'category_name': category_name}
        print(f"category_name: {category_name}")
        print ('book_data' ,book_data)
        print("---------------------------")
        books.append(book_data)
        
        
```

```python
books
```

```python
with open('book_scrap.csv','w',newline="",encoding='utf8') as f:
    writer = csv.DictWriter(f,fieldnames=['book_name','price','rating','category_name'])
    writer.writeheader()
    for book in books:
        writer.writerow(book)
```
